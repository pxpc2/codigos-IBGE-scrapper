```python
import pandas as pd

data = pd.read_csv('./ibge.csv', delimiter='\t', header=None)
```

```python
data
```

```python
municipios = []
for index, row in data.iterrows():
    columns = row[0].split(',')
    municipio = {
        "UF": columns[0],
        "COD_MUNICIPIO": columns[1],
        "NOME_MUNICIPIO": columns[2],
        "MRG": columns[3],
        "ZP": columns[4],
        "MODULO_FISCAL_ha": columns[5],
        "ZTM_IE": columns[6],
        "FMP_ha": columns[7],
        "AREAS_ESPECIAIS": columns[8]
    }
    municipios.append(municipio)

output_js_file = './output-municipios.js'
with open(output_js_file, 'w', encoding='utf-8') as js_file:
    js_file.write("export const MUNICIPIOS = " + str(municipios) + ";\n")

print(f"JavaScript file created at {output_js_file}")
```
